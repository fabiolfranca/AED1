```python
vendas_agosto = {
"04/08/2025": 2350,
"05/08/2025": 1980,
"06/08/2025": 2540,
"07/08/2025": 2100,
"08/08/2025": 3050,
"11/08/2025": 2650,
"12/08/2025": 2180,
"13/08/2025": 2760,
"14/08/2025": 2920,
"15/08/2025": 3100,
"18/08/2025": 2300,
"19/08/2025": 2430,
"20/08/2025": 2210,
"21/08/2025": 2790,
"22/08/2025": 2970,
"25/08/2025": 3100,
"26/08/2025": 2880,
"27/08/2025": 2600,
"28/08/2025": 2500,
"29/08/2025": 3400
}


vendas_agosto_valores = list(vendas_agosto.values())
vendas_agosto_keys = list(vendas_agosto.keys())


vendas_agosto_soma =  sum(vendas_agosto_valores)
vendas_agosto_media = vendas_agosto_soma / len(vendas_agosto)


semana_1 = vendas_agosto_valores[0:5]
semana_2 = vendas_agosto_valores[5:10]
semana_3 = vendas_agosto_valores[10:15]
semana_4 = vendas_agosto_valores[15:20]

print('\n' + '-' * 50 + '\n')

print('A media das vendas: ', vendas_agosto_media)
print('As somas das vendas por mês: ', vendas_agosto_soma)




semana_total_1 = sum(list(semana_1))
semana_total_2 = sum(list(semana_2))
semana_total_3 = sum(list(semana_3))
semana_total_4 = sum(list(semana_4))

print('\n' + '-' * 50 + '\n')


print('total da semana 1:', semana_total_1, 'total da semana 2:', semana_total_2, 'total semana 3:', semana_total_3, 'total da semana 4:', semana_total_4)




semana_total_1_media = semana_total_1 / len(semana_1)
semana_total_2_media = semana_total_2 / len(semana_2)
semana_total_3_media = semana_total_3 / len(semana_3)
semana_total_4_media = semana_total_4 / len(semana_4)

print('\n' + '-' * 50 + '\n')


print('total media da semana 1:', semana_total_1_media, 'total media da semana 2', semana_total_2_media, 'total media da semana 3:', semana_total_3_media, 'total media da semana 4:', semana_total_4_media) 

print('\n' + '-' * 50 + '\n')

dia_maior_venda = max(vendas_agosto_valores)
index_maior_venda = vendas_agosto_valores.index(dia_maior_venda)
data_maior_venda = vendas_agosto_keys[index_maior_venda]

print('o dia com maior venda:', data_maior_venda, 'com valor de:',dia_maior_venda, )


print('\n' + '-' * 50 + '\n')

dia_minimo_venda = min(vendas_agosto_valores)
index_minimo_venda = vendas_agosto_valores.index(dia_minimo_venda)
data_minimo_venda = vendas_agosto_keys[index_minimo_venda]

print('o dia com menor venda:', data_minimo_venda, 'com valor de:',dia_minimo_venda, )

print('\n' + '-' * 50 + '\n')

diferencial_percentual = ((dia_maior_venda - vendas_agosto_media) / dia_maior_venda) * 100 

print(f'Pico percentual: {diferencial_percentual:.2f}%')
```
