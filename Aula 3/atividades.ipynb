```python
print("\nüìä PARTE 3: TIPOS DE DADOS PRIMITIVOS EM PYTHON")
print("-" * 70)

print("\nüìã RESUMO DOS TIPOS PRIMITIVOS:")
print("+" + "-"*25 + "+" + "-"*40 + "+")
print("| {:<23} | {:<38} |".format("TIPO", "DESCRI√á√ÉO"))
print("+" + "-"*25 + "+" + "-"*40 + "+")
print("| {:<23} | {:<38} |".format("int", "N√∫meros inteiros"))
print("| {:<23} | {:<38} |".format("float", "N√∫meros decimais"))
print("| {:<23} | {:<38} |".format("str", "Texto (strings)"))
print("| {:<23} | {:<38} |".format("bool", "Valores l√≥gicos (True/False)"))
print("| {:<23} | {:<38} |".format("complex", "N√∫meros complexos"))
print("+" + "-"*25 + "+" + "-"*40 + "+")

print("\n" + "-"*70)

print("\nüî¢ TIPO INT (inteiro) - Explica√ß√£o da IA:")
tipo_int = """
O TIPO INT EM PYTHON:
‚Ä¢ Armazena n√∫meros inteiros (positivos, negativos ou zero)
‚Ä¢ N√£o tem limite de tamanho (diferente de outras linguagens)
‚Ä¢ Suporta opera√ß√µes matem√°ticas b√°sicas

PRINCIPAIS OPERA√á√ïES:
+  Adi√ß√£o          Ex: 10 + 5 = 15
-  Subtra√ß√£o       Ex: 10 - 5 = 5
*  Multiplica√ß√£o   Ex: 10 * 5 = 50
/  Divis√£o real    Ex: 10 / 3 = 3.333...
// Divis√£o inteira Ex: 10 // 3 = 3
%  M√≥dulo (resto)  Ex: 10 % 3 = 1
** Exponencia√ß√£o   Ex: 2 ** 3 = 8

ERROS COMUNS:
1. Divis√£o por zero: 10 / 0 ‚Üí ZeroDivisionError
2. Overflow: Em Python N√ÉO ocorre (tamanho ilimitado)
3. Convers√£o inv√°lida: int("abc") ‚Üí ValueError
"""

print(tipo_int)

# Demonstra√ß√£o pr√°tica
print("üß™ DEMONSTRA√á√ÉO PR√ÅTICA - TIPO INT:")
print("-" * 40)

# Opera√ß√µes b√°sicas
a = 15
b = 4

print(f"a = {a}, b = {b}")
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}")
print(f"a // b = {a // b}")
print(f"a % b = {a % b}")
print(f"a ** b = {a ** b}")

# N√∫meros grandes
print(f"\nüêò Python suporta n√∫meros enormes:")
grande_numero = 2 ** 100
print(f"2 ** 100 = {grande_numero}")
print(f"Tipo: {type(grande_numero)}")
print(f"N√∫mero de d√≠gitos: {len(str(grande_numero))}")

print("\n" + "-"*70)

print("\nüî¢ TIPO FLOAT (decimal) - Explica√ß√£o da IA:")
tipo_float = """
O TIPO FLOAT EM PYTHON:
‚Ä¢ Armazena n√∫meros com ponto decimal
‚Ä¢ Representa n√∫meros reais
‚Ä¢ Usa padr√£o IEEE 754 (precis√£o dupla - 64 bits)
‚Ä¢ Separa√ß√£o decimal com PONTO (n√£o v√≠rgula)

CARACTER√çSTICAS IMPORTANTES:
‚Ä¢ Precis√£o limitada: 15-17 d√≠gitos significativos
‚Ä¢ Nota√ß√£o cient√≠fica: 1.23e-4 = 0.000123
‚Ä¢ Underflow/overflow podem ocorrer
‚Ä¢ Problemas de arredondamento em opera√ß√µes

ERROS COMUNS:
1. Uso de v√≠rgula: 3,14 ‚Üí cria uma tupla!
2. Compara√ß√£o exata: 0.1 + 0.2 != 0.3 (problema de precis√£o)
3. Convers√µes: float("abc") ‚Üí ValueError
"""

print(tipo_float)

# Demonstra√ß√£o pr√°tica
print("üß™ DEMONSTRA√á√ÉO PR√ÅTICA - TIPO FLOAT:")
print("-" * 40)

# Opera√ß√µes com float
x = 3.14159
y = 2.71828

print(f"x = {x}, y = {y}")
print(f"x + y = {x + y}")
print(f"x * y = {x * y}")
print(f"x / y = {x / y}")

# Problema de precis√£o
print(f"\n‚ö†Ô∏è Problema de precis√£o:")
print(f"0.1 + 0.2 = {0.1 + 0.2}")
print(f"0.1 + 0.2 == 0.3 ? {0.1 + 0.2 == 0.3}")

# Solu√ß√£o para compara√ß√£o
print(f"\n‚úÖ Compara√ß√£o segura (com toler√¢ncia):")
tolerancia = 1e-10
resultado = abs((0.1 + 0.2) - 0.3) < tolerancia
print(f"abs(0.1+0.2 - 0.3) < 1e-10 ? {resultado}")

print("\n" + "-"*70)

print("\nüî¢ TIPO BOOL (booleano) - Explica√ß√£o da IA:")
tipo_bool = """
O TIPO BOOL EM PYTHON:
‚Ä¢ Armazena valores l√≥gicos: True ou False
‚Ä¢ Primeira letra MAI√öSCULA: True, False
‚Ä¢ Usado em condi√ß√µes e controle de fluxo

VALORES QUE AVALIAM COMO FALSY (False):
‚Ä¢ False
‚Ä¢ None
‚Ä¢ 0 (zero)
‚Ä¢ 0.0 (zero float)
‚Ä¢ '' (string vazia)
‚Ä¢ [] (lista vazia)
‚Ä¢ {} (dicion√°rio vazio)
‚Ä¢ () (tupla vazia)

VALORES TRUTHY (True):
‚Ä¢ Qualquer outro valor n√£o listado acima

ERROS COMUNS:
1. Letras min√∫sculas: true, false ‚Üí NameError
2. Confus√£o com strings: 'True' ‚â† True
3. Compara√ß√µes incorretas
"""

print(tipo_bool)

# Demonstra√ß√£o pr√°tica
print("üß™ DEMONSTRA√á√ÉO PR√ÅTICA - TIPO BOOL:")
print("-" * 40)

# Valores booleanos
verdadeiro = True
falso = False

print(f"verdadeiro = {verdadeiro}")
print(f"falso = {falso}")
print(f"Tipo de verdadeiro: {type(verdadeiro)}")

# Avalia√ß√£o de truthy/falsy
print(f"\nüìä Avalia√ß√£o de valores:")
valores_teste = [0, 1, 0.0, 3.14, '', 'texto', [], [1, 2], None]

for valor in valores_teste:
    resultado = bool(valor)
    print(f"bool({repr(valor):<10}) = {resultado}")

# Opera√ß√µes l√≥gicas
print(f"\nüîß Opera√ß√µes l√≥gicas:")
a = True
b = False

print(f"a and b = {a and b}")  # E
print(f"a or b = {a or b}")    # OU
print(f"not a = {not a}")      # N√ÉO

print("\n" + "-"*70)

print("\nüî¢ TIPO STR (string) - Explica√ß√£o da IA:")
tipo_str = """
O TIPO STR EM PYTHON:
‚Ä¢ Armazena sequ√™ncias de caracteres (texto)
‚Ä¢ Imut√°vel (n√£o pode modificar caracteres individuais)
‚Ä¢ Pode usar aspas simples, duplas ou triplas

FORMAS DE CRIAR STRINGS:
‚Ä¢ 'texto' ou "texto" ‚Üí linha √∫nica
‚Ä¢ '''texto''' ou \"\"\"texto\"\"\" ‚Üí m√∫ltiplas linhas
‚Ä¢ f"texto {variavel}" ‚Üí f-strings (formata√ß√£o)

PRINCIPAIS M√âTODOS:
‚Ä¢ .upper()/.lower() ‚Üí mai√∫sculas/min√∫sculas
‚Ä¢ .strip() ‚Üí remove espa√ßos
‚Ä¢ .split() ‚Üí divide string
‚Ä¢ .join() ‚Üí une strings
‚Ä¢ .find()/.replace() ‚Üí busca e substitui
‚Ä¢ .startswith()/.endswith() ‚Üí verifica in√≠cio/fim

ERROS COMUNS:
1. Aspas n√£o fechadas: 'texto ‚Üí SyntaxError
2. √çndice inv√°lido: texto[100] ‚Üí IndexError
3. Imutabilidade: texto[0] = 'x' ‚Üí TypeError
"""

print(tipo_str)

# Demonstra√ß√£o pr√°tica
print("üß™ DEMONSTRA√á√ÉO PR√ÅTICA - TIPO STR:")
print("-" * 40)

# Cria√ß√£o de strings
string1 = 'Python'
string2 = "Programa√ß√£o"
string3 = '''Texto
com m√∫ltiplas
linhas'''

print(f"string1: {string1}")
print(f"string2: {string2}")
print(f"string3: {string3}")

# F-strings (formata√ß√£o)
nome = "Carlos"
idade = 30
print(f"\nüìù F-strings:")
print(f"Nome: {nome}, Idade: {idade}")
print(f"Daqui a 5 anos: {idade + 5}")

# M√©todos comuns
texto = "  Python √© Incr√≠vel!  "
print(f"\nüîß M√©todos de string:")
print(f"Original: '{texto}'")
print(f".upper(): '{texto.upper()}'")
print(f".lower(): '{texto.lower()}'")
print(f".strip(): '{texto.strip()}'")
print(f".replace('Incr√≠vel', 'Fant√°stico'): '{texto.replace('Incr√≠vel', 'Fant√°stico')}'")
print(f".split(): {texto.strip().split()}")
print(f".find('√©'): {texto.find('√©')}")

print("\n" + "-"*70)
```
