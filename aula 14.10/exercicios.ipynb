```python
# PrÃ¡tica Integrada - AnÃ¡lise de Dados de Vendas

# 1. Dados fornecidos pelo professor (vendas diÃ¡rias de um mÃªs)
vendas_diarias = [
    1250, 980, 1750, 2100, 890,  # Semana 1
    1560, 1980, 2200, 1450, 1850,  # Semana 2
    1230, 1670, 1890, 2050, 1350,  # Semana 3
    1760, 1940, 1680, 1420, 1590,  # Semana 4
    2030, 1780, 1620, 1890, 1760,  # Semana 5 (meses com 25 dias)
    1480, 1920, 1650, 1370, 1810   # ContinuaÃ§Ã£o
]

# 2. Calcular soma total das vendas
soma_total = sum(vendas_diarias)

# Soma por semanas (5 dias por semana)
somas_semanais = []
for i in range(0, len(vendas_diarias), 5):
    semana = vendas_diarias[i:i+5]
    soma_semana = sum(semana)
    somas_semanais.append(soma_semana)

# 3. Calcular mÃ©dia diÃ¡ria
media_diaria = soma_total / len(vendas_diarias)

# 4. Encontrar valor mÃ¡ximo de vendas e seu dia
valor_maximo = max(vendas_diarias)
dia_maximo = vendas_diarias.index(valor_maximo) + 1

# 5. Encontrar valor mÃ­nimo de vendas e seu dia
valor_minimo = min(vendas_diarias)
dia_minimo = vendas_diarias.index(valor_minimo) + 1

# 6. Exibir relatÃ³rio formatado
print("=" * 60)
print("RELATÃ“RIO DE VENDAS - ANÃLISE MENSAL")
print("=" * 60)

print(f"\nðŸ“Š DADOS GERAIS:")
print(f"PerÃ­odo analisado: {len(vendas_diarias)} dias")
print(f"Total de vendas: R$ {soma_total:,.2f}")
print(f"MÃ©dia diÃ¡ria: R$ {media_diaria:,.2f}")

print(f"\nðŸ† MÃ‰TRICAS DE DESEMPENHO:")
print(f"Maior venda: R$ {valor_maximo:,.2f} (Dia {dia_maximo})")
print(f"Menor venda: R$ {valor_minimo:,.2f} (Dia {dia_minimo})")
print(f"VariaÃ§Ã£o: R$ {valor_maximo - valor_minimo:,.2f}")

print(f"\nðŸ“ˆ VENDAS POR SEMANA:")
for i, soma_semana in enumerate(somas_semanais, 1):
    dias_semana = min(5, len(vendas_diarias) - (i-1)*5)
    media_semana = soma_semana / dias_semana
    print(f"  Semana {i}: R$ {soma_semana:,.2f} (MÃ©dia: R$ {media_semana:,.2f}/dia)")

print(f"\nðŸŽ¯ ANÃLISE ADICIONAL:")

# Dias com vendas acima da mÃ©dia
dias_acima_media = [i+1 for i, venda in enumerate(vendas_diarias) if venda > media_diaria]
print(f"Dias acima da mÃ©dia: {len(dias_acima_media)} dias")
print(f"  Dias: {dias_acima_media}")

# Melhor e pior semana
melhor_semana = somas_semanais.index(max(somas_semanais)) + 1
pior_semana = somas_semanais.index(min(somas_semanais)) + 1
print(f"\nMelhor semana: Semana {melhor_semana} (R$ {max(somas_semanais):,.2f})")
print(f"Pior semana: Semana {pior_semana} (R$ {min(somas_semanais):,.2f})")

# TendÃªncia
primeira_semana = somas_semanais[0]
ultima_semana = somas_semanais[-1]
crescimento = ((ultima_semana - primeira_semana) / primeira_semana) * 100
print(f"\nðŸ“ˆ Crescimento do mÃªs: {crescimento:+.1f}%")
print(f"  (De R$ {primeira_semana:,.2f} para R$ {ultima_semana:,.2f})")

print("\n" + "=" * 60)
print("FIM DO RELATÃ“RIO")
print("=" * 60)
```
