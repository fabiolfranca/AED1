```python
#5

deposito = float(input("Digite o valor do dep√≥sito: R$ "))
taxa_juros = float(input("Digite a taxa de juros (%): "))

# C√°lculos
rendimento = deposito * (taxa_juros / 100)
valor_total = deposito + rendimento

print("\nüìä RESULTADO:")
print(f"Dep√≥sito inicial: R$ {deposito:.2f}")
print(f"Taxa de juros: {taxa_juros:.2f}%")
print(f"Rendimento: R$ {rendimento:.2f}")
print(f"Valor total ap√≥s rendimento: R$ {valor_total:.2f}")

print("\nüí° C√ÅLCULOS ADICIONAIS:")
print(f"Rendimento mensal (se taxa for anual): R$ {rendimento/12:.2f}")
print(f"Rendimento di√°rio (30 dias): R$ {rendimento/365:.2f}")

print("\nüîß VERS√ÉO PARA M√öLTIPLOS PER√çODOS:")
def calcular_juros_compostos(capital, taxa, periodos):
    montante = capital * (1 + taxa/100) ** periodos
    rendimento_total = montante - capital
    return montante, rendimento_total

# Exemplo com juros compostos
capital = 1000
taxa_anual = 10
anos = 3
montante_final, rend_total = calcular_juros_compostos(capital, taxa_anual, anos)
print(f"\nJuros compostos: R$ {capital:.2f} a {taxa_anual}% por {anos} anos")
print(f"Montante final: R$ {montante_final:.2f}")
print(f"Rendimento total: R$ {rend_total:.2f}")
```

```python
#6
base = float(input("Digite a base do tri√¢ngulo: "))
altura = float(input("Digite a altura do tri√¢ngulo: "))

# C√°lculo da √°rea
area = (base * altura) / 2

print("\nüìä RESULTADO:")
print(f"Base: {base}")
print(f"Altura: {altura}")
print(f"√Årea do tri√¢ngulo: {area:.2f}")

print("\nüí° CLASSIFICA√á√ÉO DO TRI√ÇNGULO:")
if base == altura:
    print("Tri√¢ngulo equil√°tero (todos os lados iguais)")
elif base == 0 or altura == 0:
    print("Aten√ß√£o: Base ou altura igual a zero!")
else:
    print("Tri√¢ngulo de dimens√µes diferentes")

print("\nüîß VERS√ÉO COM FUN√á√ÉO:")
def calcular_area_triangulo(b, h):
    if b <= 0 or h <= 0:
        return "Erro: Base e altura devem ser positivas"
    return (b * h) / 2

# Testando a fun√ß√£o
print("\nTestes da fun√ß√£o:")
print(f"Base 10, Altura 5: {calcular_area_triangulo(10, 5):.2f}")
print(f"Base 7.5, Altura 3.2: {calcular_area_triangulo(7.5, 3.2):.2f}")
print(f"Base 0, Altura 5: {calcular_area_triangulo(0, 5)}")

print("\nüìê C√ÅLCULO DE OUTRAS PROPRIEDADES:")
# Para tri√¢ngulo ret√¢ngulo (considerando base e altura como catetos)
if base > 0 and altura > 0:
    hipotenusa = (base**2 + altura**2) ** 0.5
    perimetro = base + altura + hipotenusa
    print(f"\nPara tri√¢ngulo ret√¢ngulo:")
    print(f"Hipotenusa: {hipotenusa:.2f}")
    print(f"Per√≠metro: {perimetro:.2f}")
```

```python
#7

raio = float(input("Digite o raio do c√≠rculo: "))
pi = 3.1415

# C√°lculo da √°rea
area = pi * (raio ** 2)

print("\nüìä RESULTADO:")
print(f"Raio: {raio}")
print(f"œÄ (pi): {pi}")
print(f"√Årea do c√≠rculo: {area:.2f}")

print("\nüí° USANDO MATH.PI PARA MAIOR PRECIS√ÉO:")
import math
area_precisa = math.pi * (raio ** 2)
print(f"Usando math.pi ({math.pi:.4f}): {area_precisa:.2f}")
print(f"Diferen√ßa: {abs(area - area_precisa):.6f}")

print("\nüîß C√ÅLCULOS RELACIONADOS:")
if raio > 0:
    diametro = 2 * raio
    circunferencia = 2 * math.pi * raio
    print(f"\nOutras medidas do c√≠rculo:")
    print(f"Di√¢metro: {diametro:.2f}")
    print(f"Circunfer√™ncia: {circunferencia:.2f}")
    print(f"Raz√£o circunfer√™ncia/di√¢metro: {circunferencia/diametro:.5f}")

print("\nüìê COMPARA√á√ÉO COM OUTRAS FIGURAS:")
# √Årea de quadrado com lado igual ao di√¢metro
lado_quadrado = 2 * raio
area_quadrado = lado_quadrado ** 2
print(f"\nQuadrado com lado igual ao di√¢metro ({lado_quadrado:.2f}):")
print(f"√Årea do quadrado: {area_quadrado:.2f}")
print(f"Raz√£o √°rea c√≠rculo/quadrado: {area_precisa/area_quadrado:.3f} (œÄ/4)")

print("\nüéØ DESAFIO EXTRA:")
print("Calcular √°rea da coroa circular (anel)")
raio_externo = float(input("\nDigite o raio externo para coroa circular: "))
if raio_externo > raio:
    area_coroa = math.pi * (raio_externo**2 - raio**2)
    print(f"√Årea da coroa circular: {area_coroa:.2f}")
else:
    print("Raio externo deve ser maior que o raio interno!")
```

```python
#8

pes = float(input("Digite a medida em p√©s: "))

# Fatores de convers√£o
PES_POR_POLEGADA = 1/12
JARDAS_POR_PE = 1/3
MILHAS_POR_JARDA = 1/1760

# Convers√µes
polegadas = pes / PES_POR_POLEGADA  # ou pes * 12
jardas = pes * JARDAS_POR_PE
milhas = jardas * MILHAS_POR_JARDA

print("\nüìä RESULTADO:")
print(f"Medida original: {pes} p√©s")
print(f"a) Em polegadas: {polegadas:.2f} in")
print(f"b) Em jardas: {jardas:.4f} jardas")
print(f"c) Em milhas: {milhas:.6f} milhas")

print("\nüí° TABELA DE CONVERS√ïES:")
print("-" * 40)
print("| Medida | Valor |")
print("-" * 40)
print(f"| P√©s    | {pes:10.2f} |")
print(f"| Polegadas | {polegadas:8.2f} |")
print(f"| Jardas | {jardas:10.4f} |")
print(f"| Milhas | {milhas:10.6f} |")
print("-" * 40)

print("\nüîß CONVERSOR UNIVERSAL:")
def converter_medidas(valor_em_pes):
    polegadas = valor_em_pes * 12
    jardas = valor_em_pes / 3
    milhas = jardas / 1760
    
    resultados = {
        'p√©s': valor_em_pes,
        'polegadas': polegadas,
        'jardas': jardas,
        'milhas': milhas
    }
    return resultados

# Testando o conversor
print("\nTestando fun√ß√£o conversora:")
medida_teste = 100
conversoes = converter_medidas(medida_teste)
for unidade, valor in conversoes.items():
    print(f"{unidade:>12}: {valor:.4f}")

print("\nüìê EXEMPLOS PR√ÅTICOS:")
print("Alturas humanas em diferentes medidas:")
alturas_pes = [5.5, 6.0, 6.5]  # 5'6", 6', 6'6"
for altura in alturas_pes:
    conv = converter_medidas(altura)
    print(f"\n{altura} p√©s = {conv['polegadas']:.1f} polegadas")
    print(f"           = {conv['jardas']:.3f} jardas")
    print(f"           = {conv['milhas']:.6f} milhas")
```

```python
#9

altura_degrau_cm = float(input("Digite a altura de cada degrau (em cm): "))
altura_desejada_metros = float(input("Digite a altura que deseja alcan√ßar (em metros): "))

# Convertendo metros para cent√≠metros
altura_desejada_cm = altura_desejada_metros * 100

# Calculando n√∫mero de degraus
degraus = altura_desejada_cm / altura_degrau_cm

print("\nüìä RESULTADO:")
print(f"Altura de cada degrau: {altura_degrau_cm} cm")
print(f"Altura desejada: {altura_desejada_metros} m ({altura_desejada_cm} cm)")
print(f"N√∫mero de degraus necess√°rios: {degraus:.1f}")

print("\nüí° CONSIDERA√á√ïES PR√ÅTICAS:")
import math
# Arredondando para cima (pois n√£o existe "meio degrau")
degraus_inteiros = math.ceil(degraus)
print(f"Voc√™ precisar√° subir {degraus_inteiros} degraus completos")

# Altura real alcan√ßada
altura_real_cm = degraus_inteiros * altura_degrau_cm
altura_real_metros = altura_real_cm / 100
diferenca = altura_real_cm - altura_desejada_cm

print(f"\nCom {degraus_inteiros} degraus, voc√™ alcan√ßar√°:")
print(f"Altura real: {altura_real_metros:.2f} m ({altura_real_cm} cm)")
print(f"Diferen√ßa em rela√ß√£o ao desejado: {diferenca:.1f} cm")

print("\nüîß VERS√ÉO COM VALIDA√á√ÉO:")
def calcular_degraus(altura_degrau, altura_desejada, unidade='metros'):
    # Converter para cm se necess√°rio
    if unidade.lower() == 'metros':
        altura_desejada_cm = altura_desejada * 100
    else:
        altura_desejada_cm = altura_desejada
    
    if altura_degrau <= 0:
        return "Erro: Altura do degrau deve ser positiva"
    
    degraus = altura_desejada_cm / altura_degrau
    degraus_inteiros = math.ceil(degraus)
    
    return degraus_inteiros, degraus

# Testando
print("\nüìã TESTES DA FUN√á√ÉO:")
testes = [
    (15, 3, 'metros'),   # degrau 15cm, altura 3m
    (18, 450, 'cm'),     # degrau 18cm, altura 450cm
    (20, 2.5, 'metros')  # degrau 20cm, altura 2.5m
]

for teste in testes:
    inteiros, decimal = calcular_degraus(*teste)
    print(f"\nDegrau {teste[0]}cm, altura {teste[1]}{teste[2][0]}:")
    print(f"  Degraus necess√°rios: {inteiros} (ideal: {decimal:.1f})")

print("\nüìê NORMAS DE SEGURAN√áA PARA ESCADAS:")
print("‚Ä¢ Altura ideal do degrau: 16-18 cm")
print("‚Ä¢ Largura m√≠nima do degrau: 28 cm")
print("‚Ä¢ N√∫mero m√°ximo de degraus consecutivos: 15")
print("‚Ä¢ Patamar obrigat√≥rio ap√≥s 15 degraus")

# Verificando se a altura do degrau est√° na norma
if 16 <= altura_degrau_cm <= 18:
    print(f"\n‚úÖ Altura do degrau ({altura_degrau_cm}cm) est√° dentro da norma")
else:
    print(f"\n‚ö†Ô∏è  Altura do degrau ({altura_degrau_cm}cm) fora da norma recomendada")
```
