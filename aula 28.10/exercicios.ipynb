```python
# Desafio 1 - Verificador de N√∫meros Pares

def eh_par(numero):
    """Verifica se um n√∫mero √© par"""
    return numero % 2 == 0

# Entrada do usu√°rio
n = int(input("Digite um n√∫mero: "))

# Verifica√ß√£o e resultado
if eh_par(n):
    print("O n√∫mero √© par!")
else:
    print("O n√∫mero √© √≠mpar!")

# Vers√£o com loop para m√∫ltiplos testes
print("\nTeste de m√∫ltiplos n√∫meros (digite 0 para sair):")
while True:
    try:
        num = int(input("\nDigite um n√∫mero (0 para sair): "))
        if num == 0:
            break
        
        if eh_par(num):
            print(f"‚úÖ {num} √© par")
        else:
            print(f"‚ùå {num} √© √≠mpar")
            
    except ValueError:
        print("Por favor, digite um n√∫mero v√°lido!")
```

```python
# Desafio 2 - Calculadora de IMC

def calcular_imc(peso, altura):
    """Calcula o √çndice de Massa Corporal"""
    if altura <= 0:
        return 0
    return peso / (altura ** 2)

def classificar_imc(imc):
    """Classifica o IMC em categorias"""
    if imc < 18.5:
        return "Abaixo do peso"
    elif 18.5 <= imc < 25:
        return "Peso normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidade"

# Entrada do usu√°rio
p = float(input("Peso (kg): "))
a = float(input("Altura (m): "))

# C√°lculo e classifica√ß√£o
imc = calcular_imc(p, a)
classificacao = classificar_imc(imc)

# Resultado
print(f"\nüìä RESULTADO:")
print(f"Peso: {p} kg")
print(f"Altura: {a} m")
print(f"IMC: {imc:.2f}")
print(f"Classifica√ß√£o: {classificacao}")

# Tabela de refer√™ncia
print("\nüìã TABELA IMC:")
print("Abaixo do peso: < 18.5")
print("Peso normal: 18.5 - 24.9")
print("Sobrepeso: 25 - 29.9")
print("Obesidade: ‚â• 30")

# Sugest√£o de peso ideal
if imc < 18.5:
    peso_ideal_min = 18.5 * (a ** 2)
    peso_ideal_max = 24.9 * (a ** 2)
    print(f"\nüí° Sugest√£o: Ganhar peso para alcan√ßar {peso_ideal_min:.1f} - {peso_ideal_max:.1f} kg")
elif imc > 25:
    peso_ideal = 22 * (a ** 2)  # IMC ideal 22
    print(f"\nüí° Sugest√£o: Perder peso para alcan√ßar aproximadamente {peso_ideal:.1f} kg")
else:
    print(f"\nüéâ Parab√©ns! Voc√™ est√° na faixa de peso saud√°vel!")
```

```python
# Desafio 3 - Simulador de Cr√©dito

def verificar_emprestimo(renda, valor_solicitado):
    """Verifica se um empr√©stimo pode ser aprovado"""
    if renda <= 0:
        return "Renda inv√°lida"
    
    percentual = (valor_solicitado / renda) * 100
    
    if percentual <= 30:
        return "Aprovado"
    elif 30 < percentual <= 50:
        return "Aprovado com restri√ß√µes"
    else:
        return "Negado"

# Entrada do usu√°rio
renda_mensal = float(input("Digite sua renda mensal (R$): "))
valor_emprestimo = float(input("Digite o valor do empr√©stimo desejado (R$): "))

# Verifica√ß√£o do empr√©stimo
resultado = verificar_emprestimo(renda_mensal, valor_emprestimo)
percentual_usado = (valor_emprestimo / renda_mensal) * 100

# Resultado
print(f"\nüìä AN√ÅLISE DO EMPR√âSTIMO:")
print(f"Renda mensal: R$ {renda_mensal:,.2f}")
print(f"Valor solicitado: R$ {valor_emprestimo:,.2f}")
print(f"Percentual da renda: {percentual_usado:.1f}%")
print(f"Situa√ß√£o: {resultado}")

# Informa√ß√µes adicionais
print("\nüí° CRIT√âRIOS DE AN√ÅLISE:")
print("‚Ä¢ At√© 30% da renda: Aprovado ‚úÖ")
print("‚Ä¢ 30% a 50% da renda: Aprovado com restri√ß√µes ‚ö†Ô∏è")
print("‚Ä¢ Acima de 50% da renda: Negado ‚ùå")

# Sugest√µes baseadas no resultado
if resultado == "Aprovado":
    print("\n‚úÖ Sua solicita√ß√£o foi aprovada! O valor est√° dentro dos limites recomendados.")
    
elif resultado == "Aprovado com restri√ß√µes":
    print("\n‚ö†Ô∏è  Cuidado! O valor solicitado est√° no limite superior.")
    valor_recomendado = renda_mensal * 0.3
    print(f"Sugest√£o: Reduza para R$ {valor_recomendado:,.2f} para aprova√ß√£o total")
    
else:  # Negado
    print("\n‚ùå Empr√©stimo negado! Valor muito alto para sua renda.")
    valor_maximo = renda_mensal * 0.3
    print(f"Sugest√£o: Solicite at√© R$ {valor_maximo:,.2f} para ter aprova√ß√£o")

# Simula√ß√£o de parcelas
if resultado != "Negado":
    print("\nüí≥ SIMULA√á√ÉO DE PARCELAS:")
    for meses in [12, 24, 36, 48]:
        parcela = valor_emprestimo / meses
        if parcela <= renda_mensal * 0.3:
            status = "‚úÖ Vi√°vel"
        elif parcela <= renda_mensal * 0.4:
            status = "‚ö†Ô∏è  No limite"
        else:
            status = "‚ùå Alto risco"
        
        print(f"{meses} meses: R$ {parcela:,.2f}/m√™s - {status}")
```
