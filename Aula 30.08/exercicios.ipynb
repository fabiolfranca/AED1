## **QUESTÕES DISSERTATIVAS**

### **QUESTÃO 13**
**Resposta:**
Ada Lovelace foi a primeira programadora da história, criando algoritmos para a Máquina Analítica de Babbage e prevendo que computadores poderiam ir além de cálculos numéricos. Alan Turing formalizou os conceitos de algoritmo e computação com a Máquina de Turing, contribuiu para decifrar o código Enigma e propôs o Teste de Turing para IA. Lovelace anteviu o potencial da computação enquanto Turing forneceu as bases matemáticas e práticas.

### **QUESTÃO 14**
**Resposta:**
A transição do ENIAC (computador massivo e caro) para o foco em software e interfaces gráficas (Microsoft e Apple) democratizou a computação. Enquanto mainframes eram para especialistas, o Windows e Macintosh tornaram os computadores acessíveis ao público geral, mudando o foco do custo do hardware para o valor do software e criando a indústria de software de consumo.

### **QUESTÃO 15**
**Resposta:**
O Teste de Turing propõe que se um avaliador humano não consegue distinguir entre uma máquina e outro humano em conversa por terminal, a máquina pode ser considerada inteligente. Isso criou uma medida operacional e tangível de inteligência artificial, focando no comportamento em vez de discussões filosóficas, servindo como benchmark para desenvolvimento de chatbots e processamento de linguagem natural.

### **QUESTÃO 16**
**Resposta:**
A ARPANET começou como projeto militar/acadêmico do DARPA em 1969, conectando universidades. A formalização do TCP/IP em 1983 permitiu a interconexão de redes diferentes, transformando-a de uma rede única em uma "rede de redes". A criação da WWW por Tim Berners-Lee em 1991 e navegadores gráficos permitiram a expansão comercial e o acesso público global, formando a Internet atual.

### **QUESTÃO 17**
**Resposta:**
A Régua de Cálculos foi um computador analógico mecânico que usava escalas logarítmicas. A Máquina Analítica de Babbage foi o primeiro design de computador digital mecânico programável. A máquina de Hollerith foi um sistema eletromecânico que usava cartões perfurados para processamento de dados em massa. Representam a evolução: analógico → digital, manual → automático, individual → em massa.

### **QUESTÃO 18**
**Resposta:**
Bit é a menor unidade de informação (0 ou 1). Byte é um conjunto de 8 bits, capaz de representar 256 valores. O byte é necessário porque um bit sozinho é insuficiente para representar caracteres humanos. Os principais códigos são: ASCII (7 bits, 128 caracteres), EBCDIC (8 bits, 256 caracteres - IBM) e Unicode (padrão moderno para todos os idiomas).

### **QUESTÃO 19**
**Resposta:**
Em notação posicional (como decimal), o valor depende da posição do dígito. Em não-posicional (como romano), o valor é fixo. Conversão de (3C9)₁₆ para decimal: 3×256 + 12×16 + 9×1 = 969. Conversão de (105)₁₀ para binário: divisões sucessivas por 2 resultam em 1101001.

### **QUESTÃO 20**
**Resposta:**
Estações de trabalho são para computação técnica individual (CAD, simulações), enquanto minicomputadores são para compartilhamento entre departamentos. A distinção desapareceu porque PCs atingiram desempenho similar a workstations e redes de PCs substituíram minicomputadores, com a evolução dos microprocessadores e sistemas operacionais tornando todos os computadores mais poderosos e acessíveis.
