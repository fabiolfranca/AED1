```python
def calculadora():
    num1 = float(input('Digite o primeiro numero: '))
    num2 = float(input('Digite o segundo numero: '))
    operacao = input('Digite a operação: + - * /: ')
    if operacao == '+':
       resultado = num1 + num2
    elif operacao == '-':
        resultado = num1 - num2
    elif operacao == '*':
       resultado = num1 * num2
    elif operacao == '/':
        resultado = num1 / num2
    else:
        print('operacao errada')
        return
    
    print(f'Resultado de {num1} {operacao} {num2} é: {resultado} ')

calculadora()
```

```python
# Desafio 2

def central_respostas(*args):
    if args:
        qtd = len(args)
        print(qtd, "Mensagens recebidas: ")
        i = 1
        for msg in args:
            print(f'Mensagem {i} recebida: {msg}')
            i = i + 1
    else:
        print('Nao tem frase')


central_respostas(
    "ola",
    "eu",
    "sou",
    "fabio"
)
```

```python
central_respostas()
```

```python
import random

def jogo():
    opcoes = ['pedra', 'papel', 'tesoura']
    nome = input('Digite seu nome: ')

    while True:
        escolha = input("Digite 'pedra', 'papel' ou 'tesoura' (ou digite 'sair'): ").lower()

        if escolha == 'sair':
            print('Jogo encerrado.')
            break

        if escolha not in opcoes:
            print('Opção inválida, tente novamente.')
            continue

        escolha_pc = random.choice(opcoes)

        print(f'\n{nome} escolheu: {escolha}')
        print(f'O computador escolheu: {escolha_pc}')

        if escolha == escolha_pc:
            print("Empate!\n")
        elif (escolha == 'pedra' and escolha_pc == 'tesoura') or \
             (escolha == 'papel' and escolha_pc == 'pedra') or \
             (escolha == 'tesoura' and escolha_pc == 'papel'):
            print("Você venceu!\n")
        else:
            print("O computador venceu!\n")

# executa o jogo
jogo()

```
