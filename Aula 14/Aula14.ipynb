```python
def operacao (operator, n1, n2):
    if operator == '+':
        return n1 + n2
    elif operator == '-':
        return n1 - n2
    elif operator == '*':
        return n1 * n2
    elif operator == '/':
        return n1 / n2

op = input('Digite a operação <+ para soma, - para subtração, * para multuplicação, / para divisão>')
n1 = float(input('Digite o primeiro número'))

if op == '/':
    while True:
        n2 = float(input('Digite o segundo número'))
        if n2 == 0:
            print('Não é possível dividir por zero! Digite o segundo número novamente!')
            continue
        break
else:
    n2 = float(input('Digite o segundo número'))

result = operacao(op, n1, n2)

print(f'{n1} {op} {n2} = {result}')
```

```python
ef potencia (n1, pot = 2):
    return n1 ** pot

# Nesse caso acima, caso a potencia não seja passada para a função, a potencia seria 2, retornando n1 ao quadrado

teste1 = potencia(4) # -> 4 elevado a 2
teste2 = potencia(4, 3) # -> 4 elevado a 3
teste3 = potencia(4, 4) # -> 4 elevado a 4

print(f'Teste 1: {teste1}\nTeste2: {teste2}\nTeste3: {teste3}')
```

```python
def calcula(op, n1, n2):
    ''' Realiza a função identificada na variável op com n1 e n2, 
    sendo n1 o primeiro número e n2 o segundo número
    Parâmetros:
    op = operação a ser realizada (+, -, * ou /)
    n1 e n2 são número inteiros
    Retorno:
    Um número inteiro'''
    if operator == '+':
        return n1 + n2
    elif operator == '-':
        return n1 - n2
    elif operator == '*':
        return n1 * n2
    elif operator == '/':
        return int(n1 / n2)
```

```python
def soma(*args):
    total = 0
    for i in args:
        total += i
    return total

# OU

def soma_com_sum(*nums):
    return sum(nums)

teste_soma_1 = soma(1, 2, 3)
teste_soma_2 = soma_com_sum(1, 2, 3, 4, 5, 6, 7, 8)

print(f'''Teste1: {teste_soma_1}
Teste2: {teste_soma_2}''')
```

```python
print('Central de Respostas\n')

def central_respostas(*frases):
    '''Recebe uma quantidade n de frases (ou nenhuma) e as reproduz, numerando-as
    Parâmetros:
    frases = uma tupla de strings
    Retorno:
    A função não tem retorno, apenas itera sobre a tupla
    ou retorna uma mensagem caso esteja vazia'''
    
    if len(frases) == 0:
        print('Nenhuma mensagem recebida. Tente novamente')
    else:
        qtd = len(frases)
        print(f'{qtd} Mensagem(ns) recebida(s). \n')
        for i in range(len(frases)):
            print(f'Mensagem {i+1}: {frases[i]}\n')


central_respostas(
"Gostaria de saber o status do meu pedido.",
"Também quero alterar meu endereço de entrega.",
"Gostaria de saber quantos pedidos foram feitos",
"Quero alterar a quantidade de produtos em um pedido"
)
```

```python
from random import choice

print('Jokenpô!\n')

player_name = input('Digite seu nome')
cpu_choices = ['pedra', 'papel', 'tesoura']
player_score = 0
cpu_score = 0
draws = 0

while True:
    player_play = input('Escolha sua jogada <"pedra", "papel" ou "tesoura">')
    cpu_play = choice(cpu_choices)
    print(f'Jogador: {player_play} x Computador: {cpu_play} ')
    if player_play == cpu_play:
        print('Empate!\n')
        draws += 1
    elif player_play == 'tesoura':
        if cpu_play == 'pedra':
            print('Vitória Computador!\n')
            cpu_score += 1
        elif cpu_play == 'papel':
            print('Vitória jogador!\n')
            player_score += 1
    elif player_play == 'papel':
        if cpu_play == 'tesoura':
            print('Vitória Computador!\n')
            cpu_score += 1
        elif cpu_play == 'pedra':
            print('Vitória jogador!\n')
            player_score += 1
    elif player_play == 'pedra':
        if cpu_play == 'papel':
            print('Vitória Computador!\n')
            cpu_score += 1
        elif cpu_play == 'tesoura':
            print('Vitória jogador!\n')
            player_score += 1
    print('Placar atual:')
    print(f'Vitórias {player_name}: {player_score}')
    print(f'Vitórias Computador: {cpu_score}')
    print(f'Empates: {draws}')

    play_again = input('Jogar novamente? <"s" para sim, "n" para não')
    if play_again == 's':
        continue
    else:
        print('Obrigado por jogar!')
        break
```
