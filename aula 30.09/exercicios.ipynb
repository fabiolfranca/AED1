```python
# Desafio 1 - Lista de N√∫meros
quantidade = int(input("Quantos n√∫meros voc√™ deseja inserir? "))

numeros = []
for i in range(quantidade):
    while True:
        try:
            numero = float(input(f"N√∫mero {i+1}: "))
            numeros.append(numero)
            break
        except ValueError:
            print("Por favor, digite um n√∫mero v√°lido!")

if numeros:
    maior = max(numeros)
    menor = min(numeros)
    soma = sum(numeros)
    media = soma / len(numeros)
    
    print(f"Lista de n√∫meros: {numeros}")
    print(f"Maior n√∫mero: {maior}")
    print(f"Menor n√∫mero: {menor}")
    print(f"Soma dos n√∫meros: {soma}")
    print(f"M√©dia dos n√∫meros: {media:.2f}")
else:
    print("Nenhum n√∫mero foi inserido!")
```

```python
# Desafio 2 - An√°lise de Frase
frase = input("Digite uma frase: ")

print(f"Frase original: '{frase}'")

# a) Total de caracteres
total_caracteres = len(frase)
print(f"a) Total de caracteres: {total_caracteres}")

# b) Total de palavras
palavras = frase.split()
total_palavras = len(palavras)
print(f"b) Total de palavras: {total_palavras}")

# c) Frase invertida
frase_invertida = frase[::-1]
print(f"c) Frase invertida: '{frase_invertida}'")
```

```python
# Desafio 3 - Jogo da Forca Simplificado
palavra_secreta = "PYTHON"
descobertas = ["_"] * len(palavra_secreta)
letras_tentadas = []
tentativas = 0

print("üéÆ JOGO DA FORCA")
print(f"Dica: √â uma linguagem de programa√ß√£o com {len(palavra_secreta)} letras")

while "_" in descobertas:
    print(f"\nPalavra: {' '.join(descobertas)}")
    print(f"Letras tentadas: {', '.join(letras_tentadas) if letras_tentadas else 'Nenhuma'}")
    
    while True:
        tentativa = input("Digite uma letra: ").upper()
        if len(tentativa) == 1 and tentativa.isalpha():
            break
        print("Por favor, digite apenas uma letra!")
    
    tentativas += 1
    
    if tentativa in letras_tentadas:
        print(f"Voc√™ j√° tentou a letra '{tentativa}'!")
        continue
    
    letras_tentadas.append(tentativa)
    
    if tentativa in palavra_secreta:
        print(f"‚úÖ Acertou! A letra '{tentativa}' est√° na palavra!")
        for i, letra in enumerate(palavra_secreta):
            if letra == tentativa:
                descobertas[i] = tentativa
    else:
        print(f"‚ùå Errou! A letra '{tentativa}' n√£o est√° na palavra.")

print(f"\nüéâ PARAB√âNS! Voc√™ descobriu a palavra '{palavra_secreta}'!")
print(f"Total de tentativas: {tentativas}")
```
